```r
library("readr")
library("tidyverse")
library ("ggplot2")
library('openxlsx')
```

```r
Tabela <- read.xlsx("RNAP-112_AS-significativos.xlsx")
Tabela %>% head
```

```r
#Selecionando só as colunas de interesse
Tab_selec <- Tabela [,c('Event','ID...1','geneSymbol','IncLevel1','IncLevel2')] 
Tab_selec$Amostra <- 'RNAP_112'
str (Tab_selec)
```

```r
#Alterar aqui para cada amostra
gene <- Tab_selec [which(Tab_selec$geneSymbol == 'MLH1'),]
gene %>% head
```

```r
#Filtrando eventos significativos 
#filter_ES <- gene %>% filter (ID...1 == 5293) # Para poucas ocorrências
filter_ES <- gene %>% filter (ID...1 %in% c(59895,59897,59899)) #Para mais de uma

filter_ES
```

```r
colnames(filter_ES) <- c('Evento','ID','Gene','PSI_CTRL','PSI_caso','Amostra')
filter_ES
```

```r
RNAP112_final <- filter_ES %>% separate_longer_delim(cols=c(PSI_CTRL),delim = ",")
str (RNAP112_final) 
```

```r
write_tsv(RNAP112_final, "RNAP112_boxplot.tsv")
```
